```python
import tika
from tika import parser 
import re
```

```python
parsed_pdf=parser.from_file("PDF_FILES/DIM evaluation report 22 Oct 2024.pdf") 
data=parsed_pdf['content']
```

```python
print(data)
```

```python
def extract_chapter_pages(data, chapter_name):
    chapter_pattern = r'^(?:\d+\.)?(?:\d+\.)?(?:\d+\.)?(?:\d+\.)?\s*([A-Z][A-Z\s]+).*?(\d+)$'
    
    chapters = []
    for line in data.split('\n'):
        line = line.strip()
        match = re.search(chapter_pattern, line)
        if match:
            title = match.group(1).strip()
            page = match.group(2)
            chapters.append((title, page))
    
    chapters.sort(key=lambda x: int(x[1]))

    target_chapter = None
    for i, (title, page) in enumerate(chapters):
        if chapter_name in title:
            target_chapter = (title, page, i)
            break
    
    if not target_chapter:
        return None, None  
    
    start_page = target_chapter[1]

    chapter_index = target_chapter[2]
    if chapter_index < len(chapters) - 1:
        end_page = int(chapters[chapter_index + 1][1]) - 1
    else:
        end_page = "END"
    
    return start_page, end_page
```

```python
start_page, end_page = extract_chapter_pages(data, "FINDINGS")
print(f"Chapter FINDINGS starts on page {start_page} and ends on page {end_page}")
```
